```python
import os
import json

def calculate_mean(file_paths):
    # 初始化各项指标的总和
    total_A_baseline = [0, 0]
    total_A_priorReal = [0, 0]
    total_A_priorGen = [0, 0]
    total_A_textReg = [0, 0]

    total_F_baseline = [0, 0]
    total_F_priorReal = [0, 0]
    total_F_priorGen = [0, 0]
    total_F_textReg = [0, 0]
    
    # 遍历所有文件
    for file_path in file_paths:
        with open(file_path, 'r') as file:
            data = json.load(file)
            
            # 提取每个文件中的指标值
            A_baseline = list(map(float, data['A']['baseline'].split(', ')))
            A_priorReal = list(map(float, data['A']['priorReal'].split(', ')))
            A_priorGen = list(map(float, data['A']['priorGen'].split(', ')))
            A_textReg = list(map(float, data['A']['textReg'].split(', ')))

            F_baseline = list(map(float, data['F']['baseline'].split(', ')))
            F_priorReal = list(map(float, data['F']['priorReal'].split(', ')))
            F_priorGen = list(map(float, data['F']['priorGen'].split(', ')))
            F_textReg = list(map(float, data['F']['textReg'].split(', ')))
            
            # 累加到总和
            total_A_baseline[0] += A_baseline[0]
            total_A_baseline[1] += A_baseline[1]
            total_A_priorReal[0] += A_priorReal[0]
            total_A_priorReal[1] += A_priorReal[1]
            total_A_priorGen[0] += A_priorGen[0]
            total_A_priorGen[1] += A_priorGen[1]
            total_A_textReg[0] += A_textReg[0]
            total_A_textReg[1] += A_textReg[1]

            total_F_baseline[0] += F_baseline[0]
            total_F_baseline[1] += F_baseline[1]
            total_F_priorReal[0] += F_priorReal[0]
            total_F_priorReal[1] += F_priorReal[1]
            total_F_priorGen[0] += F_priorGen[0]
            total_F_priorGen[1] += F_priorGen[1]
            total_F_textReg[0] += F_textReg[0]
            total_F_textReg[1] += F_textReg[1]
    
    # 计算均值
    num_files = len(file_paths)
    
    mean_A_baseline = [total_A_baseline[0] / num_files, total_A_baseline[1] / num_files]
    mean_A_priorReal = [total_A_priorReal[0] / num_files, total_A_priorReal[1] / num_files]
    mean_A_priorGen = [total_A_priorGen[0] / num_files, total_A_priorGen[1] / num_files]
    mean_A_textReg = [total_A_textReg[0] / num_files, total_A_textReg[1] / num_files]

    mean_F_baseline = [total_F_baseline[0] / num_files, total_F_baseline[1] / num_files]
    mean_F_priorReal = [total_F_priorReal[0] / num_files, total_F_priorReal[1] / num_files]
    mean_F_priorGen = [total_F_priorGen[0] / num_files, total_F_priorGen[1] / num_files]
    mean_F_textReg = [total_F_textReg[0] / num_files, total_F_textReg[1] / num_files]
    
    return (mean_A_baseline, mean_A_priorReal, mean_A_priorGen, mean_A_textReg), (mean_F_baseline, mean_F_priorReal, mean_F_priorGen, mean_F_textReg)

def get_all_json_files(folder_path):
    json_files = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith('evaluate_logs.json') and (len(root.split("/")) == 3):
                json_files.append(os.path.join(root, file))
    return json_files

# 指定文件夹路径
folder_path = 'logs/logs_continual_TR'

# 获取文件夹内所有 evaluate_logs.json 文件的路径
file_paths = get_all_json_files(folder_path)
```

```python

# 计算均值
mean_A, mean_F = calculate_mean(file_paths)

# 打印结果
print(f'Mean A Baseline: {mean_A[0]}')
print(f'Mean A Prior Real: {mean_A[1]}')
print(f'Mean A Prior Gen: {mean_A[2]}')
print(f'Mean A Text Reg: {mean_A[3]}')

print(f'Mean F Baseline: {mean_F[0]}')
print(f'Mean F Prior Real: {mean_F[1]}')
print(f'Mean F Prior Gen: {mean_F[2]}')
print(f'Mean F Text Reg: {mean_F[3]}')
```
